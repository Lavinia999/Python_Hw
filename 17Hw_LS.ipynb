```python
import json
album_ids=json.load(open('albums_ids.json','r'))
```

```python
album_ids
```

```python
TRACKS_NAMES=[]
BASE_URL='https://api.spotify.com/v1/'
```

```python
def get_tracks(album_id):
    r=requests.get(BASE_URL+'albums/'+album_id+'/tracks',headers=header)
    print(r)
    data=r.json()['items']
    return [track['name']for track in data]
```

```python
get_tracks('0F42Y0ITVGDNNRksPwgQiR')
```

```python
from concurrent import futures
```

```python
MAX_WORKERS =20  #preparation
workers = min(MAX_WORKERS,len(album_ids))
```

```python
with futures.ThreadPoolExecutor(workers) as executor:
    to_do=[]
    for album_id in album_ids:
        future = executor.submit(get_tracks, album_id)
        to_do.append(future)
    for future in futures.as_completed(to_do):
        res=future.result()
        TRACKS_NAMES.extend(res)
```
